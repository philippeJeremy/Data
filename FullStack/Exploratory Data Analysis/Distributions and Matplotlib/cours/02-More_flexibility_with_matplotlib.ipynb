## General operation of Matplotlib

### How to build a graph on Matplotlib

There are several things to consider when building a graph on Matplotlib. The diagram below should help you see them more clearly.

![](https://matplotlib.org/3.1.1/_images/anatomy.png)


Mainly, you must remember the following hierarchical elements:

* The **Figure**: A *figure* is the entire frame within which your graphic(s) is contained. You will be able to define the size of your *figure*, the number of graphics inside it, etc.

* The **Axes**: This corresponds to a graph inside your *figure*.

![](https://miro.medium.com/max/848/1*Fk80UWElsZI6f_poAv_7cQ.png)

* The **Axis**: Be careful not to confuse with the element above. The *Axis* correspond to the abscissa and ordinate of each of the graphs (or *Axes*).

* **Artist**: You may see this term in the documentation for [Matplotlib](https://matplotlib.org/3.1.1/tutorials/introductory/usage.html#interactive-example). It refers to the set of elements in the graph (including *Figure*, *Axes*, *Axis*).


### The inputs of a graphic on Matplotlib

For the moment, Matplotlib prefers to receive **numpy** objects as graphics input.

In particular, **Pandas** objects can lead to unexpected behavior. We therefore advise you to transform them via :

```python

```

## Lineplot : plt.plot()

````python
# Create random sinusoidal data
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Create the Lineplot
plt.plot(t, s)
plt.show()
````

### Some useful parameters to change the aspect

````python
# Change the color
plt.plot(t, s, color="green")
plt.show()

# Change line style
plt.plot(t, s, color="green", linestyle="--")
plt.show()
````

## Superimpose two lineplots

````python
# Create random sinusoidales datas
t = np.arange(0.0, 2.0, 0.01)
s1 = 1 + np.sin(2 * np.pi * t)
s2 = np.cos(2 * np.pi * t)

# Create Lineplot
plt.plot(t, s1, color='red')
plt.plot(t, s2, color='green')
plt.show()
````

## Histograms : plt.hist()

````python
# Distribution of a random variable x

# Create random datas
mu, sigma = 100, 15
x = mu + sigma * np.random.randn(10000)

# Create histogram
plt.hist(x, 50, facecolor='g')
plt.show()
````

````python
plt.hist?
````

### Transform the histogram into probability density
Use the argument 'density=True' to make the sum of the values in each bin equal to 1:

````python
# Create histogram
plt.hist(x, 50, facecolor='g', density=True)
plt.show()
````

### Add a probability density curve

````python
# Create random datas
mu, sigma, bins = 100, 15, 50
x = mu + sigma * np.random.randn(10000)

# Creation of the histogram
## We store in three different variables the result of plt.hist : n, bins and patches
## We will use the variable bins which corresponds to the x in our formula

n, bins, patches = plt.hist(x, bins, facecolor='g', density=True)

# Formula for density
## As seen in the figure above: x follows a normal law
y = ((1 / (np.sqrt(2 * np.pi) * sigma)) *
     np.exp(-0.5 * (1 / sigma * (bins - mu))**2))

# Add a lineplot
plt.plot(bins, y, color="r",linestyle="--")

plt.show()
````

## Pie chart : plt.pie()

````python
# Pie chart where each of the elements will be arranged counter-clockwise
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]

plt.pie(sizes, labels=labels, autopct='%1.1f%%')

plt.show()
````

````python
plt.pie?
````

### Some usefull parameters

````python
# Explode

# Pie chart where each of the elements will be arranged counter-clockwise
labels = 'Frogs', 'Hogs', 'Dogs', 'Logs'
sizes = [15, 30, 45, 10]
explode = (0, 0.1, 0, 0) # Here, only the Hogs will be highlighted.

plt.pie(sizes, explode=explode,labels=labels, autopct='%1.1f%%',
        shadow=True, startangle=90)
plt.show()
````

## Boxplots : plt.boxplot()

````python
# Creation of random data (in this case, scores on a math test)
math_test = np.random.randint(1, 20+1, 30)

# Creation of the boxplot
plt.boxplot(math_test, labels=['Math Test'])
plt.show()
````

### Some useful parameters

````python
# Change the outlier limit

# Creation of random data (in this case, scores on a math test)
math_test = np.random.randint(1, 20+1, 30)

# Add outliers
math_test = np.append(math_test, np.array([22,23,24]))

# Creation of the boxplot
plt.boxplot(math_test,
            whis=0.6,
           labels=['Math Test'])
plt.show()
````

````python
# Add the average

# Creation of random data (in this case, scores on a math test)
math_test = np.random.randint(1, 20+1, 30)

# Addition of outliers
math_test = np.append(math_test, np.array([22,23,24]))

# Creation of the boxplot
plt.boxplot(math_test,
            whis=0.6,
            showmeans=True,
            meanline=True,
            labels=['Math Test'])
plt.show()
````

## Scatter plot : plt.scatter()

````python
# Random data creation
x = np.arange(0.0, 50.0, 2.0)   # np.arange allows to create a list of numbers spaced by a defined step (here, we go from 2 to 2, 50 times)
y = x + np.random.rand(*x.shape) * 30.0 # We create an approximately linear relation in x and y

# Creation of the scatter plot
plt.scatter(x, y)
plt.show()
````

### Some useful parameters

````python
# Change the size of the markers

# Random data creation
x = np.arange(0.0, 100.0, 2.0)   # np.arange allows to create a list of numbers spaced by a defined step (here, we go from 2 to 2, 50 times)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0 # We create an approximately linear relation between x and y


# Addition of different sizes for each of the points
area = (15 * np.random.rand(50))**2  

# Creation of the scatter plot
plt.scatter(x, y, s = area)
plt.show()
````

````python
# Change the color of the markers

# Random data creation
x = np.arange(0.0, 100, 2.0)   # np.arange allows to create a list of numbers spaced by a defined step (here, we go from 2 to 2, 50 times)
y = x ** 1.3 + np.random.rand(*x.shape) * 30.0 # We create an approximately linear relation in x and y


# Addition of different sizes for each of the points
area = (15 * np.random.rand(50))**2  

# Adding a color sequence
c = ["b", "r"] * 25

# Creation of the scatter plot
plt.scatter(x, y, s=area, c = c)
plt.show()
````

## Adding information to graphics

````python
# Add titles

# Random data creation
names = ['group_a', 'group_b', 'group_c']
values = [1, 10, 100]

# Creation of the Barplot
plt.bar(names, values)
plt.title("Voici un titre") # Here we build our title
plt.show()
````

````python
# Add a legend

# Random data creation
names = ['group_a', 'group_b', 'group_c']
values_1 = [1, 10, 100]
values_2 = [7, 3, 10]

# Creation of the Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title("Voici un titre") # Create title
plt.legend(["values 1", "values 2"]) # Create legend
plt.show()
````

````python
# Add labels to abscissa & ordinate

# Data creation
names = ['group_a', 'group_b', 'group_c']
values_1 = [1, 10, 100]
values_2 = [7, 3, 10]

# Creation of the Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title(names) # Title creation
plt.legend(["values_1", "values_2"]) # Creation of legends
plt.xlabel("Abscissa axis") # Creation of labels for the abscissa
plt.ylabel("Ordinate axis") # Creation of labels for the ordinate
plt.show()
````

````python
# Add text to your graphic

# Creation of the Barplot
plt.bar(names, values_1)
plt.bar(names, values_2, bottom=values_1)
plt.title(names) # Title creation
plt.legend(["values 1", "values 2"]) # Creation of legends
plt.xlabel("Abscissa axis") # Creation of labels for the abscissa
plt.ylabel("Ordinate axis") # Creation of labels for the ordinates
plt.text(0, values_1[0]+values_2[0], names[0], horizontalalignment='center') # Adding text above the bars
plt.text(1, values_1[1]+values_2[1], names[1], horizontalalignment='center')
plt.text(2, values_1[2]+values_2[2], names[2], horizontalalignment='center')
plt.show()
````

## Another way to use matplotlib: object-oriented

````python
# Create datas
t = np.arange(0.0, 2.0, 0.01)
s = 1 + np.sin(2 * np.pi * t)

# Creation of two variables
fig, ax = plt.subplots()

# Construction of the graph
ax.plot(t, s) # Adding online data

ax.set(xlabel='time (s)', ylabel='voltage (mV)',
       title='About as simple as it gets, folks') # Here we set the title & lables for the abscissa & coordinates
plt.show()
````
