```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import statsmodels.api as sm
```

```python
df = pd.read_csv('IBM_HR_ATTRITION.csv')
```

```python
df
```

```python

df = pd.get_dummies(df, columns=["Attrition"],drop_first=True)
```

```python
sns.lmplot(x="DistanceFromHome", y="Attrition_Yes", data=df, logistic=True, scatter_kws={"alpha":.05})
```
