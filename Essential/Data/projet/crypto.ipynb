```python
df_perf = df_perf.drop(0, axis=0)
```

```python
# Compute the correlation matrix
corr = df_perf.corr()

# Generate a mask for the upper triangle
mask = np.triu(np.ones_like(corr, dtype=bool,), )

# Set up the matplotlib figure
f, ax = plt.subplots(figsize=(11, 9))

# Generate a custom diverging colormap
cmap = sns.diverging_palette(230, 20, as_cmap=True)

# Draw the heatmap with the mask and correct aspect ratio
sns.heatmap(corr, mask=mask, cmap=cmap, vmax=1, center=0, 
            square=True, linewidths=.5, cbar_kws={"shrink": .5}, annot=True, )
```

```python
features_list = [ index for i, index in enumerate(df_perf.columns) if "back" in index]


x = df_perf[features_list]
y = df_perf["for_etherium_bitcoin"]
```

```python
x_train, x_test, y_train, y_test = train_test_split(x, y, 
                                                    test_size=0.2, 
                                                    random_state=0)
```

```python
x_train
```

```python
numeric_features = [i  for i,item in enumerate(x_train.columns)]

numeric_features = [i for i in range(len(x_train.columns))]
numeric_transformer = StandardScaler()


# categorical_features = [0, 1, 2, 3] 
# categorical_transformer = OneHotEncoder()


feature_encoder = ColumnTransformer(
    transformers=[
        # ('cat', categorical_transformer, categorical_features),    
        ('num', numeric_transformer, numeric_features)
        ]
    )
```

```python
x_train = feature_encoder.fit_transform(x_train)
```

```python
regressor = RandomForestRegressor(n_estimators=30, random_state=0,max_depth=10)
regressor.fit(x_train, y_train) 
```

```python
y_train_pred = regressor.predict(x_train)
```

```python
x_test = feature_encoder.transform(x_test)
```

```python
y_test_pred = regressor.predict(x_test)
```

```python
print("R2 score on training set : ", regressor.score(x_train, y_train))
print("R2 score on test set : ", regressor.score(x_test, y_test))
```
