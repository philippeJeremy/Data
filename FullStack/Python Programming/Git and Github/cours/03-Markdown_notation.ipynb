The README.md file is what Github will display by default when you open a repository. It's important to write it well because it's what will help you understand the project and how to approach it. If you have a well done README.md file, you'll get more help on your project!

So there are a few techniques you should know about writing in a _.md_ file.


## Write headlines

Nothing could be simpler, just put a _#_ to design that your text is a title. For example :


```
# This is a title
```


The more #'s you add, the lower you go in the hierarchy of importance of the titles. For example:


```
## This is a subtitle
### This is a sub-subtitle
#### This is a sub-sub-subtitle
##### etc
#### This is the maximum
```


### Put a link

To put a link in Github, you have to proceed as follows:


```
[The text to link ](the url to redirect to)
```

For example


```
[Jedha](https://jedha.co)
```
