3. When creating a scatterplot, look at how the _LotArea_ column behaves in relation to the _SalePrices_ column.

```python
sns.relplot(x="SalePrice", y="LotArea", data=dataset)
```

4. Refine your visualization by keeping only those houses that have a _LotArea_ of less than 20,000 square feet and a price of less than $500,000.

```python
sns.relplot(x="LotArea", y="SalePrice", 
            data=dataset.loc[(dataset['LotArea'] < 20000) & (dataset['SalePrice'] < 500000), :])
```

5. When creating a point cloud, look at the relationship between the _LotFrontage_ and the _LotArea_.

```python
sns.relplot(x="LotFrontage", y="LotArea", data=dataset)
```

6. In the same way, refine your visualization by keeping only those houses that have a _LotFrontage_ of less than 200 square feet and a _LotArea_ of less than 100,000 square feet.


```python
sns.relplot(x="LotFrontage", y="LotArea", data=dataset[(dataset.LotFrontage < 200) &
                                                       (dataset.LotArea < 100000)])
```
