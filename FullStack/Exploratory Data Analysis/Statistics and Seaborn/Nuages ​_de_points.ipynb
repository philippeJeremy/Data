```python
supp_area = df[df["LotArea"] > 20000].index
```

```python
df_modif1 = df.drop(supp_area)
```

```python
supp_price = df_modif1[df_modif1["SalePrice"] > 500000].index
```

```python
df_modif1.drop(supp_price, inplace=True)
```

```python
sns.relplot(x='LotArea', y='SalePrice', data=df_modif1)
```
