```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df_expo = np.random.exponential(scale=1, size=5000, )
```

```python
df_expo 
```

```python
sns.distplot(df_expo)
```

```python
df_expo = np.reshape(df_expo, (1000, 5))
```

```python
df_expo
```

```python
df2 = []
for i in df_expo:
    d = i.mean()
    df2.append(d)
```

```python
sns.distplot(df2)
```

```python
np.mean(df2)
```

```python
np.std(df2)
```

```python
df_normal = np.random.normal(scale=0.4356682851255858, loc=0.9927839429357179, size=1000, )
```

```python
df_normal
```

```python
sns.distplot(df2)
sns.distplot(df_normal, hist=False, color='green')
```

```python
sns.distplot(df2)
sns.distplot(df_normal2, hist=False, color='green')
```
