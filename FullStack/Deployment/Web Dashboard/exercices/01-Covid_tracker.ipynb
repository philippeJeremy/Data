Let's build our first dashboard using streamlit. 

For this first exercise, we got inspiration from the great [Covid Tracker app](https://covidtracker.fr/) built by [Guillaume Rozier](https://fr.wikipedia.org/wiki/Guillaume_Rozier). 

Your goal will be to build a dashboard that follow each covid wave throughout the year. To do so, we are going to use data from the European Centre for Disease Prevention and Control (ECDC). 

* [Download data here](https://www.ecdc.europa.eu/en/publications-data/data-daily-new-cases-covid-19-eueea-country)


Here is an example of what you can create: 

* [Example dashboard](https://covid-tracker-fs-streamlit-app.herokuapp.com/)
