
## Agile project management

There are several management methods that have developed over the years. One that is very popular is the _Agile method, which is very useful, fast and flexible. Especially when applied to technical teams. So let's look at the main principles.


## Agile VS Waterfall

Agile methods consist in being able to quickly develop a product or a product feature. The idea is to be able to be flexible and adapt quickly to market needs.

You will therefore define with your team an MVP or Minimum Viable Product. This means that you will include the strict minimum possible in your product so that it has a minimum of attractiveness. You will then test this product by launching it on the market, collect feedback from your target and finally you will "iterate" on the product, i.e. improve it to make it more attractive. Once you've created your second version, you'll send it back to the market and so on, until you get the perfect product.

These methods are very fashionable because they work extremely well for services and software, which is an important part of today's economy.  

However, they have their limits when it comes to industrial products where production cycles are very long and do not allow as much flexibility. This is why we will use Waterfall methods instead, where the idea is to prepare an action plan that is executed from start to finish. This method has less flexibility but it allows you to set a course for the whole team in the development of a project.


## The SCRUM method

We talked about the main principles of project management. Now, let's focus on one of the agile methods: _SCRUM._ You will very often have to manage projects or be part of projects managed by the SCRUM method. We will therefore explain it in more detail.


### The actors of the method

First of all, the _SCRUM framework distinguishes three roles within a _SCRUM team:


1. A _Product Owner_ who will carry the vision of the product. That is to say, he will explain how he sees the finished product.
2. A _Development Team_ that will take care of building the product.
3. A _Scrum Master_ who will guarantee that the _SCRUM_ methodology is applied.

The _SCRUM_ teams can be from 3 to 9 people. Beyond that, you lose agility.


## Sprint

At the beginning of the SCRUM method, there is the Sprint. That is to say, we define the elements that the development team will carry out during a defined period of time. Sprints are often over 30 days.

The elements we are talking about are first of all part of the _Product Backlog_, i.e. the elements put on hold for development. The Product Owner will define the elements to be developed in priority and assign resources to them. Once these are defined, the development team gets to work.


### Daily Stand Up

Every day, the development team takes 15 minutes to explain what they are going to work on and what they have finished. This allows everyone to stay on top of things and move quickly through the day.


### Sprint review

Once the Sprint is over, the development team will be going through all the features they have managed to develop. We also collect feedback from users and the product owner and prepare the next sprint.


### Sprint Retrospective

Finally, we have one last meeting which is a retrospective of the sprint. The purpose of this meeting is to determine how to improve the team's productivity. We're going to do an introspection and see what worked and what didn't work so well after the sprint experience.
