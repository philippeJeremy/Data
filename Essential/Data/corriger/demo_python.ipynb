```python
# Import of libraries 
import pandas as pd 
import numpy as np 
import seaborn as sns 
```

```python
# Build and Visualize Dataset
df = pd.read_csv("km_traveled_all_transportations.csv").iloc[:,1:]
df
```

```python
# Percentage of conversions 
df.groupby("type_of_transport").mean()
```

```python
# Visualisation of catplot version : conversion rate with 95% confidence level
sns.catplot(x="type_of_transport", y="Km_traveled", data = df, kind="bar", ci=50)
```

```python
df = pd.read_csv("km_traveled_per_transport.csv").iloc[:,1:]
df
```

```python
# with confidence level at 0.9
t_value=1.96
sigma = df.std()
n = len(df)
E = t_value * sigma / np.sqrt(n)

df.mean().plot(yerr=E, kind="bar")
```
